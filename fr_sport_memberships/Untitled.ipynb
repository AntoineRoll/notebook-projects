```python
import pandas as pd

clubs_df = pd.read_csv("clubs-data-2021.csv", sep=";")
clubs_df
```

```python
clubs_par_departements = clubs_df.groupby(['Fédération', 'Département'], as_index=False).agg({"Commune": "count"})
clubs_par_departements
```

```python
clubs_par_departements[clubs_par_departements.Fédération=='FF de Rugby'].sort_values('Commune', ascending=False)
```

```python
clubs_df[(clubs_df.Département=='38') & (clubs_df.Fédération=='FF de Rugby')]
```

```python
clubs_departement = clubs_df[clubs_df.Département=='14']
clubs_departement.groupby('Fédération', as_index=False).agg(
    {"Clubs": "sum"}
).sort_values('Clubs', ascending=False).iloc[:50].reset_index()
```

```python
licencies_df = pd.read_csv('lic-data-2021.csv', sep=';')
licencies_sports = licencies_df.groupby(['Fédération', 'Département'], as_index=False).agg(
    {"Total": "sum"}
)
licencies_sports
```

```python
licencies_sports.sort_values("Total", ascending=False).iloc[:50]
```

```python
licencies_sports.groupby('Fédération', as_index=False).sum('Total').sort_values('Total', ascending=False).iloc[:50]
```

```python
top_federations = licencies_sports.groupby('Fédération', as_index=False).sum('Total').sort_values('Total', ascending=False).iloc[:50]

for federation in top_federations.Fédération:
    print(f'["{federation}"],')
```
